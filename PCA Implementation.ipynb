```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


from sklearn.datasets import load_breast_cancer
```

```python
cancer= load_breast_cancer()
```

```python
cancer.keys()
classifications =cancer['target'].reshape(569,1)
print(classifications)
## CREATE NEW DATA FRAME WITH THE FEATURE AS EACH COLUMN

df= pd.DataFrame(cancer['data'], columns=cancer['feature_names'])

```

```python
## Now we standardize (standard scaling)

from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()
scaled_Data= scaler.fit_transform(df)
print(scaled_Data)

```

```python
## Data has now been standardized. Next we compute the covariance matrix

from sklearn.decomposition import PCA

pca= PCA(n_components=2)
principal_components= pca.fit_transform(scaled_Data)
print(scaled_Data.shape)
principal_components.shape

##By comparing the shape of the initial scaled data and the principal component variable we can see
##that the dimensions of have been reduced by the Principal Component Analysis
```

```python
##Now we check the percentage of data maintained in our PCs
print(principal_components)
                    
```

```python
pca_fit= pca.fit(scaled_Data)
print(pca_fit.explained_variance_ratio_)
```

```python
##PLOTS
classes= cancer.target.reshape(569,1)

plt.figure(figsize=(8,6))
plt.scatter(principal_components[:,0], principal_components[:,1], c=classes)
```

```python
from sklearn.datasets import load_digits
digits= load_digits()
print(digits.keys())
data= digits.data
```

```python
pca1= PCA(n_components=2)
dig_pca= pca1.fit_transform(data)
print(data.shape)
dig_pca.shape

target= digits.target.reshape(1797,1)
print(target)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(dig_pca[:,0], dig_pca[:,1], c= target, alpha=0.5, cmap=plt.cm.get_cmap('prism',10))
plt.colorbar()
```

```python
def plot_digits(data):
    #create a figure with subplots, 4 rows, 8 columns, with figsize 10x4, with no x or y ticks
    #grid_space determines the space between the various subplots
    
    fig, axes = plt.subplots(4, 10, figsize=(10, 4),
                             subplot_kw={'xticks':[], 'yticks':[]},
                             gridspec_kw=dict(hspace=0.2, wspace=0.2))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                  cmap='binary', interpolation='nearest',
                  clim=(0, 16))

plot_digits(data)
```
