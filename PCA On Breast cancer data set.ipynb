```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler


```

```python
##Using breast cancer data set for classification (malignent or benign)

breast= load_breast_cancer()
breast_data= breast.data
breast_labels= breast.target

##target  is used to fetch labels
```

```python
labels = np.reshape(breast_labels,(569,1))
final_breast_data = np.concatenate([breast_data,labels],axis=1)
print(final_breast_data)
```

```python
breast_dataset= pd.DataFrame(final_breast_data)
features= breast.feature_names
print(features)
```

```python
print(final_breast_data)
```

```python
features_labels= np.append(features, 'label')
print(features_labels)
```

```python
breast_data= pd.DataFrame(final_breast_data)
breast_data.columns= features_labels
print(breast_data)
```

```python
breast_data.head()
```

```python
breast_data['label'].replace(0,'Benign', inplace=True)
breast_data['label'].replace(1, 'Malignant', inplace= True)
breast_data.tail()
```

```python
breast_dataset.columns= features_labels

```

```python
breast_dataset['label'].replace(0,'Benign', inplace=True)
breast_dataset['label'].replace(1, 'Malignant', inplace= True)

```

```python
#Standardization
x=  breast_dataset.loc[:,features].values
x= StandardScaler().fit_transform(x)
print(x)
x.shape
```

```python
np.mean(x), np.std(x)
```

```python
feature_columns= ['feature'+ str(i) for i in range(x.shape[1])]
normalised_breast_data=  pd.DataFrame(x, columns= feature_columns)
print(normalised_breast_data)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_breast= PCA(n_components=2)
principal_components_breast = pca_breast.fit_transform(normalised_breast_data)


pca_bdf= pd.DataFrame(data= principal_components_breast, columns=['principal component 1', 'principal component 2'])
print(pca_bdf)
```

```python
print(pca_breast.explained_variance_ratio_)
print('The first 2 principal components capture {}% of the original data'.format(np.sum(pca_breast.explained_variance_ratio_)*100))
```
